### Low Pass Filter (Exponential Smoothing)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define Low Pass Filter
def low_pass_filter(x, est, alpha=0.2):
    return (1-alpha)*est + alpha*x
```

```python
# Set Random Seed
np.random.seed(42)

# Generate Data
minimum = 0
maximum = 30
std = 4
num = 100

gt = data = np.linspace(minimum, maximum, num=num)
noise = np.random.normal(0, std, num)
data = gt + noise

# Get Estimates
alpha = 0.2
ests = []
for i, x in enumerate(data):
    if i == 0:
        est = x
    else:
        est = low_pass_filter(x, est, alpha=alpha)
    ests.append(est)
```

```python
# Graph
plt.plot(data, 'o', color='k', alpha=0.5, label='Data')
plt.plot(gt, lw=5, color='b', label='Ground Truth')
plt.plot(ests, lw=5, color='r', label='Estimation')
plt.legend()
plt.xlim(0, num)
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```
