### Average Filter

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define Average Filter
def average_filter(x, est, n=1):
    alpha = 1/n
    return (1-alpha)*est + alpha*x
```

```python
# Set Random Seed
np.random.seed(42)

# Generate Data (MEAN=16, STD=4)
mean = 16
std = 4
num = 100

data = np.random.normal(mean, std, num)

# Get Estimates
ests = []
for i, x in enumerate(data):
    if i == 0:
        est = x
    else:
        est = average_filter(x, est, n=i+1)
    ests.append(est)
```

```python
# Graph
plt.plot(data, 'o', color='k', alpha=0.5, label='Data')
plt.plot([mean]*num, lw=5, color='b', label='Ground Truth')
plt.plot(ests, lw=5, color='r', label='Estimation')
plt.legend()
plt.xlim(0, num)
plt.ylim(0, 30)
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```
